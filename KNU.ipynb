```
import konlpy
from konlpy.tag import Kkma
from konlpy.utils import pprint
import json
import jpype
from tqdm import tqdm
import random

print(jpype.isJVMStarted()) #return False:not running or 0:running

class KnuSL():

	def data_list(wordname):
		with open('./SentiWord_info.json', encoding='utf-8-sig', mode='r') as f:
			data = json.load(f)
		result = ['None','None']	
		for i in range(0, len(data)):
			if data[i]['word'] == wordname:
				result.pop()
				result.pop()
				result.append(data[i]['word_root'])
				result.append(data[i]['polarity'])	
		
		r_word = result[0]
		s_word = result[1]
							
		#print('어근 : ' + r_word)
		#print('극성 : ' + s_word)		
		
		
		return r_word, s_word
```

```
def read_data(file_path):
    with open(file_path, "r", encoding="utf8") as inFile:
        lines = inFile.readlines()

    datas = []
    for index, line in enumerate(tqdm(lines, desc="read_data")):
        # 입력 문장을 \t으로 분리
        pieces = line.strip().split("\t")

        # 데이터의 형태가 올바른지 체크
        assert len(pieces) == 3

        if(index == 0):
            continue

        id, sequence, label = pieces[0], pieces[1], int(pieces[2])
        datas.append((id, sequence, label))

    return datas
```

```
kkma = Kkma()
ksl = KnuSL

datas = read_data('./ratings_train_final.txt')

datas_list = random.sample(datas, 100000)

correct = 0
total = 0

for data in datas_list:
  print('['+str(total)+'] ', end='\t')
  print(data[1])

  words = konlpy.tag.Okt().pos(data[1], norm=True, stem=True)

  print(words)
  sum = 0
  predict = 0

  for word in words:
    #print(word)
    data_list = ksl.data_list(word[0])
    #print(data_list)
  
    try: sum += int(data_list[1])
    except: continue

  if sum>=0: predict = 1

  if predict==data[2]: correct += 1

  print('predict='+str(predict), end='\t')
  print('answer='+str(data[2]))
  print()

  total += 1

result = correct/total

print('correctness= '+str(result)+'%')
```
